# How do I create plots in pandas?

```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
# kolom pertama dijadikan index dengan index_col dan tipe data diubah menjadi timestamp dengan parse_dates
df_air = pd.read_csv('/content/drive/MyDrive/dataset/pandas_tutorial/air_quality_no2.csv', index_col=0, parse_dates=True)
df_air
```

```python
df_air.describe()
```

```python
df_air.info()
```

The usage of the index_col and parse_dates parameters of the read_csv function to define the first (0th) column as index of the resulting DataFrame and convert the dates in the column to Timestamp objects, respectively.

```python
# visual data
# x label adalah index_col/dates
df_air.plot()
plt.show()
```

```python
# visual one column
df_air['station_paris'].plot()
plt.show()
```

```python
# plot two column
df_air[['station_paris', 'station_london']].plot()
plt.show()
```

```python
# scatter plot
df_air.plot.scatter(x="station_london", y="station_paris", alpha=0.5)
plt.show()
```

```python
# boxplot
df_air.plot.box()
plt.show()
```

I want each of the columns in a separate subplot.

```python
# no subplot
df_air.plot.area(figsize=(12, 4))
plt.show()
```

```python
# with subplot
df_air.plot.area(figsize=(12, 4), subplots=True)
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(12, 4))        # Create an empty Matplotlib Figure and Axes
df_air.plot.area(ax=axs)                        # Use pandas to put the area plot on the prepared Figure/Axes
axs.set_ylabel("NO$_2$ concentration")          # Do any Matplotlib customization you like
axs.set_xlabel("Waktu")                         # Do any Matplotlib customization you like
fig.savefig("no2_concentrations.png")           # Save the Figure/Axes using the existing Matplotlib method.
plt.show()                                      # Display the plot
```

# How to calculate summary statistics

```python
import pandas as pd
```

```python
titanic = pd.read_csv('/content/drive/MyDrive/dataset/pandas_tutorial/titanic.csv')
titanic
```

```python
titanic.describe()
```

### Aggregating statistics

What is the average age of the Titanic passengers?

```python
titanic['Age'].mean()
```

What is the median age and ticket fare price of the Titanic passengers?

```python
titanic[['Age', 'Fare']].median()
```

```python
male = titanic[(titanic['Age'] > 35) & (titanic['Sex'].isin(['male']))]
male
```

```python
titanic[['Age', 'Fare']].mode()
```

```python
titanic[['Age', 'Fare']].describe()
```

Instead of the predefined statistics, specific combinations of aggregating statistics for given columns can be defined using the DataFrame.agg() method:

```python
titanic.agg(
    {
        # dic => key:value
        "Age": ["min", "max", "median", "skew"],
        "Fare": ["min", "max", "median", "mean"],
    }
)
```

# Aggregating statistics grouped by category

What is the average age for male versus female Titanic passengers?

```python
# [tabel yang diambil atau ditampilkan].groupby(dikelompokkan berdasarkan kolom apa) => fungsi statistik
titanic[['Age', 'Sex', 'SibSp', 'Parch']].groupby('Sex').mean()
```

```python
# jika ingin menampilkan summary stat dari semua kolom, tambahkan numeric_only=True agar hanya menghitung pada kolom dengan value numerik saja
titanic.groupby('Sex').mean(numeric_only=True)
```

```python
# ingin menampilkan summary dari beberapa kolom pilihan saja
# cara ini lebih fleksibel
titanic.groupby("Sex")[["Age", "SibSp", "Parch"]].mean()
```

What is the mean ticket fare price for each of the sex and cabin class combinations?

```python
# groupby 2 kolom kategori
titanic.groupby(["Sex", "Pclass"])[["Fare"]].mean()
```
