```python
import pandas as pd
```

```python
air_quality_no2 = pd.read_csv('/content/drive/MyDrive/dataset/pandas_tutorial/air_quality_no2_long.csv', parse_dates=True)
air_quality_no2
```

```python
air_quality_no2 = air_quality_no2[["date.utc", "location",
                                   "parameter", "value"]]
air_quality_no2
```

```python
air_quality_pm25 = pd.read_csv("/content/drive/MyDrive/dataset/pandas_tutorial/air_quality_pm25_long.csv",
                               parse_dates=True)
```

```python
air_quality_pm25 = air_quality_pm25[["date.utc", "location",
                                     "parameter", "value"]]


air_quality_pm25
```

# How to combine data from multiple tables

Combine NO2 and pm25 using concat

```python
# gunakan pd.concat
# arti axis=0  maka digabungkan secara vertikal/baris
air_quality = pd.concat([air_quality_pm25, air_quality_no2], axis=0)

air_quality
```

```python
# cek shape
print('Shape dari ``air_quality_pm25`` table: ', air_quality_pm25.shape)

print('Shape dari ``air_quality_no2`` table: ', air_quality_no2.shape)

print('Shape dari gabungan ``air_quality`` table: ', air_quality.shape)
```

```python
# sorting tables dari gabungan table
air_quality = air_quality.sort_values("date.utc")

air_quality.head()
```

# Join tables using a common identifier

```python
stations_coord = pd.read_csv("/content/drive/MyDrive/dataset/pandas_tutorial/air_quality_stations.csv")

stations_coord.head()
```

```python
air_quality.head()
```

JOIN TABLE USING pd.merge

```python
# jika dari dua tabel memiliki nama kolom yang sama
air_quality = pd.merge(air_quality, stations_coord, how="left", on="location")

air_quality.head()
```

```python
air_quality_parameters = pd.read_csv("/content/drive/MyDrive/dataset/pandas_tutorial/air_quality_parameters.csv")

air_quality_parameters.head()
```

```python
# jika dari dua tabel memiliki kolom kunci yang berbeda, gunakan left_on dan right_on
air_quality = pd.merge(air_quality, air_quality_parameters,
                       how='left', left_on='parameter', right_on='id')


air_quality.head()
```
